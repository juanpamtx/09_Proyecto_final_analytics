Importamos las librerías necesarias para hacer un eda preliminar

```python
import pandas as pd


import sys
sys.path.append('../')

```

Cargamos los datos o datasets 

```python
customers = pd.read_csv('../data/01_raw/customers.csv')
behavior = pd.read_csv('../data/01_raw/behavior.csv')
products = pd.read_csv('../data/01_raw/products.csv')
transactions = pd.read_csv('../data/01_raw/transactions.csv')


```

```python
for name, df in {
    "customers": customers,
    "behavior": behavior,
    "products": products,
    "transactions": transactions
}.items():
    print(f"\n{name.upper()} → {df.shape}")
    print(df.head())
    print(df.info())
    print('-'*20)
```

Una vez analizados los datos que tenemos, podemos decir lo siguiente:
- **Customer** --> primary key: **'customer_id'**. Se combina con *'behavior', 'transaction'*

- **Behavior** --> primary key: **'customer_id'**. Se combina con *'customer', 'transaction'*

- **Products** --> primary key: **'product_id'**. Se combina con *'transaction'*

- **Transaction** --> primary key: **'customer_id', 'product_id'**. Se combina con *'ambos'*
