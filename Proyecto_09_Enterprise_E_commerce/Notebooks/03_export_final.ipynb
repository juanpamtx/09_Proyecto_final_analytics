1. Importamos librerías

```python
import pandas as pd
import os
import sys

project_root = os.path.abspath(os.path.join(".."))
sys.path.append(project_root)
```

2. Cargamos los datasets limpios

```python
ruta_processed = "../data/02_processed/"

behavior = pd.read_csv(ruta_processed + "behavior_clean.csv")
customers = pd.read_csv(ruta_processed + "customers_clean.csv")
products = pd.read_csv(ruta_processed + "products_clean.csv")
transactions = pd.read_csv(ruta_processed + "transactions_clean.csv")
```

3. Unimos los datasets con un "merge"

```python
## Unimos "transactions" con "products" para obtener el nombre del producto

df = transactions.merge(products, on="product_id", how="left")
```

```python
## Añadimos "customers" para obtener el nombre del cliente

df = df.merge(customers, on="customer_id", how="left")
```

```python
## Finalement añadimos "behavior" para obtener la fecha de la compra

df = df.merge(behavior, on="customer_id", how="left")
```

4. Comprobamos el resultado final 

```python
print(f'Nuestro nuevo archivo de datos "df" tiene un total de {df.shape[0]} filas y {df.shape[1]} columnas')

display(df.head())
```

5. Guardamos el daset final en la ruta data/03_final para su posterior análisis.

```python
ruta_final = "../data/03_final/"
os.makedirs(ruta_final, exist_ok=True)

df.to_csv(ruta_final + "ecommerce_master_dataset.csv", index=False)

print("Dataset final exportado correctamente.")
```
