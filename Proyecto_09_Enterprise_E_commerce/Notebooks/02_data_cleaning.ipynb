1. Importamos librería y funciones generales

```python
# Tratamiento de datos

import pandas as pd
pd.set_option('display.max_columns', None)
import sys
import os

# Añadimos ruta al proyecto para poder importar módulos
project_root = os.path.abspath(os.path.join(".."))
sys.path.append(project_root)

# ----------

# Funciones personalizadas

from src.utils.cleaning_utils import (
    limpiar_behavior,
    limpiar_customers,
    limpiar_products,
    limpiar_transactions
)

```

2. Cargamos los datasets desde la ruta data/01_raw

```python

ruta_raw = "../data/01_raw/"

behavior = pd.read_csv(ruta_raw + "behavior.csv")
customers = pd.read_csv(ruta_raw + "customers.csv")
products = pd.read_csv(ruta_raw + "products.csv")
transactions = pd.read_csv(ruta_raw + "transactions.csv")

print("Datasets cargados correctamente.")

```

3. Vemos los datasets originales antes de hacer los cambios que hemos planteado anteriormente.

```python
print("=== BEFORE CLEANING: Behavior ===")
display(behavior.head())

print("=== BEFORE CLEANING: Customers ===")
display(customers.head())

print("=== BEFORE CLEANING: Products ===")
display(products.head())

print("=== BEFORE CLEANING: Transactions ===")
display(transactions.head())
```

4. Aplicamos las funciones de limpieza y mostramos los cambios

Behavior

```python
print("=== Cleaning Behavior ===")

behavior_clean = limpiar_behavior(behavior)

print("Nuevas columnas creadas:")
[k for k in behavior_clean.columns if "binaria" in k or "yes_no" in k]

display(behavior_clean.head())
```

Customers

```python
print("=== Cleaning Customers ===")

customers_clean = limpiar_customers(customers)

print("Nuevas columnas creadas:")
[k for k in customers_clean.columns if "binaria" in k or "yes_no" in k]

display(customers_clean.head())
```

Products

```python
print("=== Cleaning Products ===")

products_clean = limpiar_products(products)

display(products_clean.head())
```

Transactions

```python
print("=== Cleaning Transactions ===")

transactions_clean = limpiar_transactions(transactions)

print("Nuevas columnas creadas:")
[k for k in transactions_clean.columns if "binaria" in k or "yes_no" in k]

display(transactions_clean.head())
```

Una vez hechos los cambios, procedemos a guardar los nuevos datasets en la carpeta data/02_processed

```python
ruta_processed = "../data/02_processed/"

behavior_clean.to_csv(ruta_processed + "behavior_clean.csv", index=False)
customers_clean.to_csv(ruta_processed + "customers_clean.csv", index=False)
products_clean.to_csv(ruta_processed + "products_clean.csv", index=False)
transactions_clean.to_csv(ruta_processed + "transactions_clean.csv", index=False)

print("Archivos guardados en data/02_processed/")
```
